Deskripsi Dataset
Dataset ini berasal dari penjualan sebuah perusahaan supermarket yang beroperasi di beberapa cabang di kota Myanmar . Dalam beberapa bulan terakhir, perusahaan ini telah mencatat data penjualan dari tiga cabang yang berbeda. Pertumbuhan supermarket dan kompetisi di pasar yang semakin ketat mendorong pentingnya analisis data penjualan. Dengan menggunakan metode analitik prediktif, dataset ini memberikan peluang untuk memahami perilaku pelanggan dan kinerja penjualan.
Informasi Atribut Dataset ini terdiri dari berbagai atribut yang mencerminkan detail transaksi penjualan. Berikut adalah deskripsi dari setiap atribut:
1.	Invoice ID:

o	Deskripsi: Nomor identifikasi unik untuk setiap slip penjualan yang dihasilkan secara otomatis oleh sistem.

o	Tipe: Kategorikal
2.	Branch:

o	Deskripsi: Cabang supermarket tempat transaksi terjadi. Terdapat tiga cabang yang diidentifikasi dengan huruf A, B, dan C.

o	Tipe: Kategorikal
3.	City:

o	Deskripsi: Lokasi dari masing-masing cabang supermarket.

o	Tipe: Kategorikal
4.	Customer Type:

o	Deskripsi: Tipe pelanggan yang mencatat apakah mereka menggunakan kartu anggota (Member) atau tidak (Normal).

o	Tipe: Kategorikal
5.	Gender:

o	Deskripsi: Jenis kelamin pelanggan, yang mencakup kategori "Male" dan "Female".

o	Tipe: Kategorikal

6.	Product Line:

o	Deskripsi: Kategorisasi umum dari item yang dijual, seperti Aksesori Elektronik, Aksesori Fashion, Makanan dan Minuman, Kesehatan dan Kecantikan, Rumah dan Gaya Hidup, serta Olahraga dan Perjalanan.

o	Tipe: Kategorikal
7.	Unit Price:

o	Deskripsi: Harga per unit untuk masing-masing produk yang dijual, dinyatakan dalam dolar ($).

o	Tipe: Numerik

8.	Quantity:

o	Deskripsi: Jumlah produk yang dibeli oleh pelanggan dalam satu transaksi.

o	Tipe: Numerik

9.	Tax 5%:

o	Deskripsi: Biaya pajak sebesar 5% dari total pembelian yang dikenakan kepada pelanggan.

o	Tipe: Numerik
10.	Total:

o	Deskripsi: Total harga pembelian termasuk pajak.

o	Tipe: Numerik

11.	Date:

o	Deskripsi: Tanggal transaksi berlangsung. Data ini mencakup periode dari Januari 2019 hingga Maret 2019.

o	Tipe: Kategorikal

12.	Time:

o	Deskripsi: Waktu transaksi terjadi, berkisar antara pukul 10.00 hingga 21.00.

o	Tipe: Kategorikal

13.	Payment:

o	Deskripsi: Metode pembayaran yang digunakan pelanggan untuk melakukan pembelian. Terdapat tiga metode pembayaran: Tunai, Kartu Kredit, dan E-wallet.

o	Tipe: Kategorikal

14.	COGS (Cost of Goods Sold):

o	Deskripsi: Biaya barang yang dijual, yang menunjukkan total biaya langsung dari produksi barang yang dijual.

o	Tipe: Numerik

15.	Gross Margin Percentage:

o	Deskripsi: Persentase margin kotor yang menunjukkan seberapa besar keuntungan yang diperoleh setelah dikurangi biaya barang.

o	Tipe: Numerik

16.	Gross Income:

o	Deskripsi: Pendapatan kotor yang dihasilkan dari penjualan barang, sebelum dikurangi biaya.

o	Tipe: Numerik

17.	Rating:

o	Deskripsi: Penilaian dari pelanggan tentang pengalaman berbelanja mereka, diberikan pada skala 1 hingga 10.

o	Tipe: Numerik

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy.stats import skew


# Kolom numerik dan kategorikal yang relevan berdasarkan informasi yang diberikan
numerical_columns = ['Unit price', 'Quantity', 'Tax 5%', 'Total', 'cogs', 'gross margin percentage', 'gross income', 'Rating']
categorical_columns = ['Invoice ID', 'Branch', 'City', 'Customer type', 'Gender', 'Product line', 'Date', 'Time', 'Payment']


dark_pink = '#C71585'


# 1. Visualisasi untuk variabel numerik
for col in numerical_columns:
    print(f"\nAnalisis distribusi untuk kolom: {col}")

    # Visualisasi Histogram untuk distribusi
    plt.figure(figsize=(8, 6))
    sns.histplot(df[col], kde=True, color=dark_pink)  # Mengganti warna menjadi pink gelap
    plt.title(f'Distribusi {col}')
    plt.xlabel(col)
    plt.ylabel('Frekuensi')
    plt.show()

  # 2. Menghitung pemusatan dan penyebaran
    mean_value = df[col].mean()
    median_value = df[col].median()
    std_dev = df[col].std()

    print(f'Rata-rata: {mean_value}')
    print(f'Median: {median_value}')
    print(f'Deviasi Standar: {std_dev}')

  # 3. Menghitung Skewness
    data_skewness = skew(df[col].dropna())  # Menghindari nilai NaN
    print(f'Skewness: {data_skewness}')

  # 4. Menentukan kategori distribusi berdasarkan skewness
    if data_skewness < -1:
        distribution = "Distribusi sangat miring ke kiri"
    elif -1 <= data_skewness < 0:
        distribution = "Distribusi miring ke kiri"
    elif 0 < data_skewness <= 1:
        distribution = "Distribusi miring ke kanan"
    else:
        distribution = "Distribusi sangat miring ke kanan"

    print(f'Kategori distribusi: {distribution}')

# Visualisasi untuk variabel kategorikal
for col in categorical_columns:
    print(f"\nAnalisis distribusi untuk kolom: {col}")

    # Visualisasi Count Plot untuk kategori
    plt.figure(figsize=(10, 6))
    sns.countplot(x=df[col], order=df[col].value_counts().index, color=dark_pink)  # Mengganti warna menjadi pink gelap
    plt.title(f'Distribusi {col}')
    plt.xlabel(col)
    plt.ylabel('Frekuensi')
    plt.xticks(rotation=45, ha='right')
    plt.show()
```

IDENTIFIKASI DISTRIBUSI SETIAP VARIABEL PADA DATA SUPERMARKET SALES DI NEGARA MYANMAR

1. Invoice ID (Distribusi Uniform atau Nominal)


	Variabel ini adalah pengenal unik dari setiap transaksi. Invoice ID bisa dianggap mengikuti distribusi uniform atau nominal, karena setiap transaksi memiliki ID yang unik dan tidak ada pengulangan.
2. Branch (Distribusi Binomial)


	Branch adalah variabel kategorikal dengan tiga cabang berbeda (misalnya, A, B, C). Ini bisa dianggap sebagai distribusi binomial di mana setiap cabang merupakan kategori dari distribusi tersebut.
3. City (Distribusi Binomial atau Nominal)


	Variabel City mencakup beberapa kota, dan masing-masing kota adalah kategori. Ini bisa dianggap mengikuti distribusi binomial atau nominal, tergantung pada jumlah kota.
4. Customer type (Distribusi Bernaulli)


	Customer type biasanya memiliki dua kategori, misalnya, "Member" dan "Non-member". Dengan dua kategori, ini mengikuti distribusi Bernaulli.
5. Gender (Distribusi Bernaulli)

	Variabel Gender biasanya memiliki dua nilai, "Male" dan "Female", sehingga ini jelas mengikuti distribusi Bernaulli.
6. Product line (Distribusi Binomial)

	Product line mencakup beberapa jenis produk (misalnya, Electronics, Health, Clothing), sehingga ini dapat dianggap mengikuti distribusi binomial dengan beberapa pilihan.
7. Unit price (Distribusi Normal atau Eksponensial)

	Harga satuan suatu produk kemungkinan besar mengikuti distribusi normal, dengan sebagian besar harga mendekati nilai rata-rata dan menyebar secara simetris. Namun, jika data harga menunjukkan penurunan yang cepat di ujung tinggi atau rendah, ini mungkin mengikuti distribusi eksponensial.
8. Quantity (Distribusi Poisson atau Eksponensial)

	Jumlah barang yang dibeli dalam satu transaksi seringkali mengikuti distribusi Poisson, karena ini adalah jumlah kejadian diskrit dalam suatu rentang waktu. Jika distribusinya lebih merata, bisa juga dianggap eksponensial.
9. Tax 5% (Distribusi Linear atau Eksponensial)

	Pajak adalah 5% dari total pembelian. Karena ini merupakan nilai yang diturunkan secara proporsional dari total, ini mengikuti distribusi yang linear atau mungkin eksponensial, tergantung pada pola pembelian.
10. Total (Distribusi Normal atau Eksponensial)

	Total belanja kemungkinan besar mengikuti distribusi normal jika mayoritas nilai berpusat di rata-rata. Jika ada banyak transaksi kecil dan hanya sedikit transaksi besar, ini bisa mengikuti distribusi eksponensial.
11. Date (Distribusi Uniform)

	Tanggal transaksi biasanya mengikuti distribusi uniform, karena setiap tanggal dalam jangka waktu tertentu memiliki probabilitas yang sama untuk terjadinya transaksi.
12. Time (Distribusi Uniform atau Eksponensial)

	Waktu transaksi juga bisa dianggap mengikuti distribusi uniform, di mana semua waktu dalam sehari memiliki probabilitas yang sama, atau eksponensial jika ada waktu-waktu tertentu dengan lonjakan transaksi yang tinggi.
13. Payment (Distribusi Binomial)

	Metode pembayaran (misalnya, Cash, Credit Card, E-wallet) adalah variabel kategorikal dengan beberapa pilihan. Ini mengikuti distribusi binomial.
14. cogs (Cost of Goods Sold) (Distribusi Normal)

	Biaya barang yang dijual (cogs) cenderung mengikuti distribusi normal jika biaya rata-rata menyebar secara simetris di sekitar nilai pusat.
15. Gross margin percentage (Distribusi Normal)

	Persentase margin kotor juga biasanya mengikuti distribusi normal, dengan sebagian besar nilai berada di sekitar rata-rata.
16. Gross income (Distribusi Normal atau Eksponensial)

	Pendapatan kotor (gross income) bisa mengikuti distribusi normal jika mayoritas transaksi menghasilkan pendapatan yang rata-rata. Jika ada distribusi dengan sedikit transaksi besar, ini bisa menjadi eksponensial.
17. Rating (Distribusi Normal)

	Penilaian (rating) dari pelanggan cenderung mengikuti distribusi normal jika sebagian besar nilai rating mendekati rata-rata, seperti nilai tengah 3-4 pada skala 1-5.
